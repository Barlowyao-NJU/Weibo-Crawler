```python
# coding=utf-8
import xlrd
import requests
import re
import time
import os
import csv
import sys
import json
import pandas as pd
from config import headers,Cookie
    
initial_url='https://m.weibo.cn/api/container/getIndex?uid=1871802012&t=0&luicode=10000011&lfid=100103type%3D1%26q%3D%E6%B7%B1%E5%9C%B3%E5%A4%A9%E6%B0%94&type=uid&value=1871802012&containerid=1076031871802012'
next_url='https://m.weibo.cn/api/container/getIndex?uid=1871802012&t=0&luicode=10000011&lfid=100103type%3D1%26q%3D%E6%B7%B1%E5%9C%B3%E5%A4%A9%E6%B0%94&type=uid&value=1871802012&containerid=1076031871802012&since_id={}'
continue_url = initial_url 
data_Dict = {'data':[]}
pageCount=0

def Get_CHN(text):
    dr = re.compile(r'<[^>]+>',re.S)
    dd = dr.sub('',text)
    data_Dict['data'].append(dd)
    write_to_excel(data_Dict)

def get_longtext_page(id):
    extendurl='https://m.weibo.cn/statuses/extend?id={0}'.format(str(id))
    r = requests.get(extendurl, headers=headers)
    if r.status_code == 200:
        return r.json()
    
def parse_longtext_page(data):
    if data['ok']==1:
        Get_CHN(data['data']['longTextContent'])

def write_to_excel(content):
    pdData = pd.DataFrame(content)
    pdData.to_excel("深圳天气_20210421_100pages.xlsx",encoding='utf_8_sig')
        
def get_page(url): #请求页面
    try:
        requests.DEFAULT_RETRIES = 5  # 增加重试连接次数
        s = requests.session()
        s.keep_alive = False  # 关闭多余连接
        r = requests.get(url, headers=headers,cookies=Cookie)
        if r.status_code == 200:
            return r.json()
        else:
            if r.status_code == 200:
                return r.json()
    except requests.ConnectionError as e:
        print('error', e.args)

        
def get_data(data): #解析页面返回的json数据
    if data['ok']==1:
        since_id = data.get("data").get("cardlistInfo").get("since_id")
        if since_id != 0:
            for cardGroup in data['data']['cards']:
                if cardGroup['card_type']==9: 
                    if cardGroup['mblog']['isLongText'] == True:
                        parse_longtext_page(get_longtext_page(cardGroup['mblog']['id']))  
                    elif cardGroup['mblog']['isLongText'] == False:
                        Get_CHN(cardGroup['mblog']['text'])
            print(since_id)
            global pageCount
            pageCount+=1

            global next_url # 通过since_id生成下一个页面的url
            continue_url = next_url.format(str(since_id))
            print(continue_url)
            print('====={0}====='.format(pageCount))
            if pageCount<=100:
                time.sleep(5)
                print("now to next page")
                datafornextpage=get_page(continue_url) #访问下一个页面
                get_data(datafornextpage) #函数内递归调用
            else: 
                print("page to 100, mission complete")
                
        else:
            print("no since id")
    else:
        print('error')
        
 
if __name__ == "__main__":   
        jsondata = get_page(initial_url)
        get_data(jsondata)
```
